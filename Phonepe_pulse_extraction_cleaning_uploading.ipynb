```python
import json
import os
import pandas as pd
from pymongo import MongoClient

```

# AGGREGATED  <--->  TRANSACTION

```python
#TO GET THE DATA-FRAME OF AGGREGATED <--> TRANSACTION
path1=r"C:/Users/Faith/PycharmProjects/phonepe/data/aggregated/transaction/country/india/state/"

col1 = {'State': [], 'Year': [], 'Quater': [], 'Transaction_type': [], 'Transaction_count': [],
        'Transaction_amount': []}

Agg_trans_state_list = os.listdir(path1)
for i in Agg_trans_state_list:
    p_i = path1 + i + "/"
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = p_i + j + "/"
        Agg_yr_list = os.listdir(p_j)
        #print(Agg_yr_list)

        for k in Agg_yr_list:
            p_k = p_j + k
            #print(p_k)
            Data = open(p_k)
            A = json.load(Data)
            
            try:
                
                for z in A['data']['transactionData']:
                    Name = z['name']
                    count = z['paymentInstruments'][0]['count']
                    amount = z['paymentInstruments'][0]['amount']
                    col1['Transaction_type'].append(Name)
                    col1['Transaction_count'].append(count)
                    col1['Transaction_amount'].append(amount)
                    col1['State'].append(i)
                    col1['Year'].append(j)
                    col1['Quater'].append(int(k.strip('.json')))
            
            except:
                
                pass
#print(type(col1))
df_aggregated_transaction = pd.DataFrame(col1)
```

```python
df_aggregated_transaction
```

# converting transaction amount in crores

```python
df_aggregated_transaction['Transaction_amount'] = (df_aggregated_transaction['Transaction_amount'].astype(float)/10000000).astype(str) + 'Cr'
```

```python
df_aggregated_transaction.isna().sum()
```

# AGGREGATED  <-->  USER

```python
# TO GET THE DATA-FRAME OF AGGREGATED <--> USER
path2 = r"C:/Users/Faith/PycharmProjects/phonepe/data/aggregated/user/country/india/state/"
Agg_user_state_list = os.listdir(path2)

col2 = {'State': [], 'Year': [], 'Quater': [], 'Brands': [], 'Count': [],
        'Percentage': []}
for i in Agg_user_state_list:
    p_i = path2 + i + "/"
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = p_i + j + "/"
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = p_j + k
            # print(p_k)
            Data = open(p_k)
            B = json.load(Data)
            
            try:
                
                for w in B["data"]["usersByDevice"]:
                    brand_name = w["brand"]
                    count_ = w["count"]
                    ALL_percentage = w["percentage"]
                    col2["Brands"].append(brand_name)
                    col2["Count"].append(count_)
                    col2["Percentage"].append(ALL_percentage)
                    col2["State"].append(i)
                    col2["Year"].append(j)
                    col2["Quater"].append(int(k.strip('.json')))
            
            except:
                
                pass
df_aggregated_user = pd.DataFrame(col2)
```

# TOP  <-->  TRANSACTION

```python
# TO GET THE DATA-FRAME OF TOP <--> TRANSACTION
path5=r"C:/Users/Faith/PycharmProjects/phonepe/data/top/transaction/country/india/state/"
top_trans_state_list = os.listdir(path5)
col5 = {'State': [], 'Year': [], 'Quater': [], 'Pincode':[], 'Transaction_count': [],
        'Transaction_amount': []}
col6= {'State': [], 'Year': [], 'Quater': [], 'District': [],  'Transaction_count': [],
        'Transaction_amount': []}
for i in top_trans_state_list:
    p_i = path5 + i + "/"
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = p_i + j + "/"
        Agg_yr_list = os.listdir(p_j)
        
        for k in Agg_yr_list:
            p_k = p_j + k
            #print(p_k)
            Data = open(p_k)
            E = json.load(Data)
            
            try:
                
                for a in E['data']['pincodes']:
                    
                    pincode = a['entityName']
                    #print(pincode)
                    count = a['metric']['count']
                    amount = a['metric']['amount']
                    col5['Pincode'].append(pincode)
                    col5['Transaction_count'].append(count)
                    col5['Transaction_amount'].append(amount)
                    col5['State'].append(i)
                    col5['Year'].append(j)
                    col5['Quater'].append(int(k.strip('.json')))
                    
                for b in E['data']['districts']:
                    count = b['metric']['count']
                    amount = b['metric']['amount']
                    district=b['entityName']
                    col6['District'].append(district)
                    col6['Transaction_count'].append(count)
                    col6['Transaction_amount'].append(amount)
                    col6['State'].append(i)
                    col6['Year'].append(j)
                    col6['Quater'].append(int(k.strip('.json')))
                
            except:
                
                pass
            
df_top_transaction_pincode = pd.DataFrame(col5)
df_top_transaction_district = pd.DataFrame(col6)                
            
            
```

# TOP  <-->  USER

```python
# TO GET THE DATA-FRAME OF TOP <--> USER
path6=r"C:/Users/Faith/PycharmProjects/phonepe/data/top/user/country/india/state/"
top_user_state_list = os.listdir(path6)
col7 = {'State': [], 'Year': [], 'Quater': [], 'Pincode':[],
        'RegisteredUser': []}
col8 = {'State': [], 'Year': [], 'Quater': [], 'District': [],
        'RegisteredUser': []}
for i in top_user_state_list:
    p_i = path6 + i + "/"
    Agg_yr = os.listdir(p_i)

    for j in Agg_yr:
        p_j = p_i + j + "/"
        Agg_yr_list = os.listdir(p_j)

        for k in Agg_yr_list:
            p_k = p_j + k
            #print(p_k)
            Data = open(p_k)
            F = json.load(Data)
            
            try:
                
                for a in F['data']['pincodes']:
                    pincode=a['name']
                    #print(pincode)
                    registeredUser = a['registeredUsers']
                    col7['Pincode'].append(pincode)
                    col7['RegisteredUser'].append(registeredUser)
                    col7['State'].append(i)
                    col7['Year'].append(j)
                    col7['Quater'].append(int(k.strip('.json')))
                    
                for b in F['data']['districts']:
                    district=b['name']
                    registeredUser = b['registeredUsers']
                    col8['District'].append(district)
                    col8['RegisteredUser'].append(registeredUser)
                    col8['State'].append(i)
                    col8['Year'].append(j)
                    col8['Quater'].append(int(k.strip('.json')))
                    
            except:
                
                pass
            
df_top_user_pincode = pd.DataFrame(col7)
df_top_user_district = pd.DataFrame(col8)          
                
                    
            
```
